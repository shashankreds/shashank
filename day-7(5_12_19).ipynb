```
# even sum using functions
def even_sum(n):
  sum=0
  for i in range(1,n+1):
    if i%2 ==0:
      sum+=i
  return sum
even_sum(30)
```

```

#finding factorial of a number using functions
def factorial(n):
  f=1
  for i in range (1,n+1):
    f=f*i
  return f
print(factorial(5))
print(factorial(10))
```
